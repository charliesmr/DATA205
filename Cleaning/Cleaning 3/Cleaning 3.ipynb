Fisrt I import the pandas library and read the "clean_deer" csv file.

I display the first 5 rows of the data.

```python
import pandas as pd

filepath = "clean_deer.csv"

data = pd.read_csv(filepath)

data.head()
```

I then use .info() to get some details about the data columns.

```python
data.info()
```

Next I filter for years from 2006 to 2010 and assign it to a second dataframe. I display the first 5 rows of data.

I do this so I can analyze the deer population with the 5-year census estimates in my population dataset.

```python
data2 = data[(data['Year'] >= 2006) & (data['Year'] <= 2010)]
data2.head()
```

I then group by park name and sum the number of deer for each park. I again display the first 5 rows.

```python
data3 = data2.groupby(['ParkName'])['DeerNum'].sum()
data3 = data3.reset_index()
data3.head()
```

I create another column with the years as my previous mutating removed it.

```python
list_year = ['2006-2010', '2006-2010', '2006-2010']

data3['Years'] = list_year
data3.head()
```

I then repeat the process for the years 2011 to 2015.

```python
data4 = data[(data['Year'] >= 2011) & (data['Year'] <= 2015)]
data4.head()
```

I group by park name and sum the number of deer for each park.

```python
data5 = data4.groupby(['ParkName'])['DeerNum'].sum()
data5 = data5.reset_index()
data5.head()
```

Like before I have to add another column for the years.

```python
list_year2 = ['2011-2015', '2011-2015', '2011-2015']

data5['Years'] = list_year2
data5.head()
```

For the third 5-year census population in my population dataset I filter for 2016 to 2020.

```python
data6 = data[(data['Year'] >= 2016) & (data['Year'] <= 2020)]
data6.head()
```

Again I group by park name and sum the number of deer.

```python
data7 = data6.groupby(['ParkName'])['DeerNum'].sum()
data7 = data7.reset_index()
data7.head()
```

I add the year column.

```python
list_year3 = ['2016-2020', '2016-2020', '2016-2020']

data7['Years'] = list_year3
data7.head()
```

Next I set the indices for all three dataframes so I can combine them.

```python
df1 = pd.DataFrame(data3)
df2 = pd.DataFrame(data5)
df3 = pd.DataFrame(data7)

df1.index = [1, 2, 3]
df2.index = [4, 5, 6]
df3.index = [7, 8, 9]
```

I use .concat from the pandas library to combine all three dataframes into one.

```python
frames = [df1, df2, df3]

final_data = pd.concat(frames)
final_data.head(10)
```

Finally, I create a .csv file with the deer population data.

```python
final_data.to_csv('deer_pop_data.csv', index = False)
```
