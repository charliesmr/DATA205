First I import the pandas library and read the census data .csv file.
I rename the columns using .columns and display the first 5 rows of the dataframe.

```python
import pandas as pd

filepath = "Census_Data.csv"

data = pd.read_csv(filepath)
data.columns = ['county', 'total_pop', 'years']
data.head()
```

Next I remove the , in the total population column so I can use it as a integer variable.

```python
for i in data.index:
  total_pop_str = data.loc[i, 'total_pop']
  total_pop = int(total_pop_str.replace(',', ''))
  data.loc[i, 'total_pop'] = total_pop
data.head()
```

I use .convert_dtype to convert the total pop column into an integer instead of a string.

```python
data = data.convert_dtypes()
data.info()
```

Next I will combine Christian County and Greene County into one. The reason for this is because unlike Arkansas Post and Pea Ridge parks, Wilson's Creek is surrounded by two counties instead of one and to use this population data with my original dataset, I need to convert it the counties into one value.

I start by filter for the 5-year census estimate from 2006 to 2010. I will combine the two counties year by year instead of doing it all at once as it's easier this way.

```python
data2 = data[(data['years'] == "2006-2010")]
data2.head()
```

I set the index of the rows I want to combine and use .loc to get the row data. I then add the elements of the rows together into "combined_row".

I use .loc to replace row 10 with the combined row and drop row 11.

```python
row10_index = 10
row11_index = 11

row10 = data2.loc[row10_index]
row11 = data2.loc[row11_index]

combined_row = row10 + row11
combined_row

data2.loc[row10_index] = combined_row

df1 = data2.drop(row11_index)

df1
```

Next I edit the names of the string values.

```python
df1['county'] = df1['county'].replace({'Christian County, MissouriGreene County, Missouri': 'Christian/Greene Counties, Missouri'})
df1['years'] = df1['years'].replace({'2006-20102006-2010': '2006-2010'})
df1.head()
```

I repeat the process for 2011 to 2015.

```python
data3 = data[(data['years'] == "2011-2015")]
data3.head()
```

```python
row6_index = 6
row7_index = 7

row6 = data3.loc[row6_index]
row7 = data3.loc[row7_index]

combined_row = row6 + row7
combined_row

data3.loc[row6_index] = combined_row

df2 = data3.drop(row7_index)

df2
```

```python
df2['county'] = df2['county'].replace({'Christian County, MissouriGreene County, Missouri': 'Christian/Greene Counties, Missouri'})
df2['years'] = df2['years'].replace({'2011-20152011-2015': '2011-2015'})
df2.head()
```

Again I do the same process for 2016 to 2020.

```python
data4 = data[(data['years'] == "2016-2020")]
data4.head()
```

```python
row2_index = 2
row3_index = 3

row2 = data4.loc[row2_index]
row3 = data4.loc[row3_index]

combined_row = row2 + row3
combined_row

data4.loc[row2_index] = combined_row

df3 = data4.drop(row3_index)

df3
```

```python
df3['county'] = df3['county'].replace({'Christian County, MissouriGreene County, Missouri': 'Christian/Greene Counties, Missouri'})
df3['years'] = df3['years'].replace({'2016-20202016-2020': '2016-2020'})
df3.head()
```

Next I set the index of the three dataframes so I can combine them.

```python
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
df3 = pd.DataFrame(df3)

df1.index = [1, 2, 3]
df2.index = [4, 5, 6]
df3.index = [7, 8, 9]
```

I use .concat from the pandas library to combine the dataframes into one.

```python
frames = [df1, df2, df3]

final_data = pd.concat(frames)
final_data.head(10)
```

Finally, I create a .csv file with the population data so I can use it for future analysis.

```python
final_data.to_csv('final_pop_data.csv', index = False)
```
